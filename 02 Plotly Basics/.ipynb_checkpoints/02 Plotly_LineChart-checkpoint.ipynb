```python
# importing libraries
import numpy as np
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
np.random.seed(56)
```

```python
x_values = np.linspace(0, 1, 100)
y_values = np.random.randn(100)
```

```python
trace0 = go.Scatter(x=x_values, y=y_values+5, 
                  mode='markers', name='markers')
```

```python
trace1 = go.Scatter(x=x_values, y=y_values, mode='lines', name='mylines')
```

```python
trace2 = go.Scatter(x=x_values, y=y_values-5, mode='lines+markers', name='my fav')
```

```python
data = [trace0, trace1, trace2]
```

```python
layout = go.Layout(title='Line Charts')
```

```python
fig= go.Figure(data=data, layout=layout)
```

```python
pyo.plot(fig, filename='linechart.html' )
```

<h2>Line Charts on a Data Set</h2>

```python
# importing libraries
import plotly.offline as pyo
import plotly.graph_objs as go
import pandas as pd
```

```python
df = pd.read_csv('../data/nst-est2017-alldata.csv')
```

```python
df.head()
```

```python
# filtering division
df_newengland = df[df['DIVISION']=='1']
```

```python
df_newengland.head()
```

```python
# Setting the name of the divion to be the index
df_newengland.set_index('NAME', inplace=True)
```

```python
# grab columns with populstion data using list comprehension
list_of_pop_col = [col for col in df_newengland.columns if col.startswith('POP')]
```

```python
df_newengland = df_newengland[list_of_pop_col]
```

```python
df_newengland
```

```python
#  build a list comprehension to build traces
data = [go.Scatter(x=df_newengland.columns,
                   y=df_newengland.loc[name], 
                   mode='lines', 
                   name=name) for name in df_newengland.index]
```

```python
pyo.plot(data, filename='population_line.html')
```

Using 2010YumaAZ.cs to develop a line chart

Plots seven days worth of temps data on one graph

```python
# importing libraries
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
df = pd.read_csv('../data/2010YumaAZ.csv')
```

```python
df.head()
```

```python
days = ['TUESDAY', 'WEDNESDAY', 'THURSDAY', 'FRIDAY', 'SATURDAY', 'SUNDAY', 'MONDAY']
```

```python
# using a for loop (or list comprehension to create traces for the data list)
data = []
```

```python
for day in days:
    trace = go.Scatter(
        x=df['LST_TIME', 
        y=df[df['DAY'==day]]['T_HR_AVG'], 
        mode='lines', name=day)
    data.append(trace)
```
